## Validation Results

```python
import matplotlib.pyplot as plt

print(
    f'\n##################################\n\n'
    f'      Validations Results\n\n'
    f'##################################\n\n'
    f'Curva de pontuação F1 (F1_curve.png): Esta curva representa a pontuação F1 em vários limiares. A interpretação desta curva pode oferecer informações sobre o equilíbrio do modelo entre falsos positivos e falsos negativos em diferentes limiares.\n\n'
    f'Curva de precisão-recuperação (PR_curve.png): Uma visualização integral para qualquer problema de classificação, esta curva mostra os compromissos entre a precisão e a recuperação em limites variados. Torna-se especialmente significativa quando lida com classes desequilibradas.\n\n'
    f'Curva de precisão (P_curve.png): Representa graficamente os valores de precisão em diferentes limiares. Esta curva ajuda a compreender como a precisão varia à medida que o limiar muda.\n\n'
    f'Curva de recuperação (R_curve.png): De forma correspondente, este gráfico ilustra a forma como os valores de recordação mudam em diferentes limiares.\n\n'
    f'Matriz de confusão (confusion_matrix.png): A matriz de confusão fornece uma visão detalhada dos resultados, mostrando as contagens de verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos para cada classe.\n\n'
    f'Matriz de confusão normalizada (confusion_matrix_normalized.png): Esta visualização é uma versão normalizada da matriz de confusão. Representa os dados em proporções em vez de contagens brutas. Esse formato simplifica a comparação do desempenho entre as classes.\n\n'
    f'Etiquetas de lotes de validação (val_batchX_labels.jpg): Estas imagens mostram os rótulos verdadeiros para lotes distintos do conjunto de dados de validação. Fornecem uma imagem clara do que são os objectos e as suas respectivas localizações de acordo com o conjunto de dados.\n\n'
    f'Previsões do lote de validação (val_batchX_pred.jpg): Em contraste com as imagens de etiquetas, estas imagens apresentam as previsões efectuadas pelo modelo YOLOv8 para os respectivos lotes. Ao compará-las com as imagens das etiquetas, podes avaliar facilmente a eficácia do modelo na deteção e classificação visual dos objectos.\n\n'
)


# Supondo que você já tem os nomes dos arquivos dos gráficos
f1_curve = f'{dataset.location}/RESULTS_VAL/15/F1_curve.png'
pr_curve = f'{dataset.location}/RESULTS_VAL/15/PR_curve.png'
p_curve = f'{dataset.location}/RESULTS_VAL/15/P_curve.png'
r_curve = f'{dataset.location}/RESULTS_VAL/15/R_curve.png'
confusion_matrix = f'{dataset.location}/RESULTS_VAL/15/confusion_matrix.png'
confusion_matrix_normalized = f'{dataset.location}/RESULTS_VAL/15/confusion_matrix_normalized.png'

# Lista de legendas para cada gráfico
legend_labels = ['Curva de pontuação F1', 'Curva de precisão-recuperação', 'Curva de precisão',
                 'Curva de recuperação', 'Matriz de confusão', 'Matriz de confusão normalizada']

# Lista de nomes dos arquivos dos gráficos
graph_files = [f1_curve, pr_curve, p_curve, r_curve, confusion_matrix, confusion_matrix_normalized]

# Para cada gráfico, plotar e adicionar a legenda correspondente
for i in range(len(graph_files)):
    plt.figure()
    # Plotar o gráfico
    plt.imshow(plt.imread(graph_files[i]))
    # Adicionar a legenda
    plt.title(legend_labels[i])
    plt.axis('off')  # Desativar os eixos
    plt.show()

print('Baixo mAP: Indica que o modelo pode necessitar de aperfeiçoamentos gerais.\n'
      'IoU baixo: O modelo pode estar a ter dificuldades em localizar os objectos com precisão. A utilização de diferentes métodos de caixas delimitadoras pode ajudar.\n'
      'Baixa precisão: O modelo pode estar a detetar demasiados objectos inexistentes. Ajustar os limites de confiança pode reduzir isso.\n'
      'Recorda pouco: O modelo pode estar a perder objectos reais. Melhorar a extração de características ou utilizar mais dados pode ajudar.\n'
      'Pontuação F1 desequilibrada: Existe uma disparidade entre a precisão e a recuperação.\n'
      'AP específico da turma: As pontuações baixas aqui podem destacar as disciplinas em que o modelo tem dificuldades.')
```
